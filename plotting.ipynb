```python
import json
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (8, 6)

# As classificações foram feitas para
# 4, 8, 12, 16, 20, 24, 28, 32, 36, 40, 44, 48, 52, 56, 60, 64 vizinhos
num_neighbours = [4, 8, 12, 16, 20, 24, 28, 32, 36, 40, 44, 48, 52, 56, 60, 64]
```

## australian.dat

```python
with open('data/australian_stats.json','r') as ausf:
    data = json.loads(ausf.read())
```

### Classes

```python
list(data[0]['precision'].keys())
```

### Acurácia
---

```python
arr = []
for i in range(16):
    arr.append(data[i]['accuracy'])

for i, v in enumerate(arr):
    plt.text(i, v, "%.2f" %v, ha='right', va='baseline', fontsize='small')

plt.plot(range(len(num_neighbours)), arr, 'o:')
plt.xticks(range(len(num_neighbours)), num_neighbours)
plt.xlabel('número de vizinhos computados')
plt.ylabel('acurácia (%)')
plt.title('Acurácia')
plt.savefig('australian_acuracia.png')
```

```python
arr = []
for i in range(16):
    arr.append(data[i]['precision']['0.0']['precision'])

for i, v in enumerate(arr):
    plt.text(i, v, "%.2f" %v, ha='right', va='baseline', fontsize='small')

plt.plot(range(len(num_neighbours)), arr, 'o:')
plt.xticks(range(len(num_neighbours)), num_neighbours)
plt.xlabel('número de vizinhos computados')
plt.ylabel('precisão (%)')
plt.title('Precisão da classificação "0"')
plt.savefig('australian_precisao_0.png')
```
